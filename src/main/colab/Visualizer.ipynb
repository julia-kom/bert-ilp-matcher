```python
file_path = path + "Cologne.pnml"

net, initial_marking, final_marking = pnml_importer.apply(path + "uni/Cologne.pnml")

gviz = pn_visualizer.apply(net, initial_marking, final_marking)
pn_visualizer.view(gviz)

pnml_exporter.apply(net, initial_marking, "petri.pnml")

pnml_exporter.apply(
    net, initial_marking, "petri_final.pnml", final_marking=final_marking
)

```
